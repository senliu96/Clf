```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import random
random.seed(33)
```

```python
plt.figure(figsize=(12,12))
x,y = make_blobs(n_samples=500,random_state=27,cluster_std=0.7)
plt.subplot(211)
plt.scatter(x[:,0],x[:,1],c=y)
plt.title("input")
plt.show()
x[2]
```
